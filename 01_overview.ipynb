# S&P500, Nasdaq 100, and Dow Jones

I'll work on the companies listed in S&P500, Nasdaq 100, and Dow Jones. I need the symbols of the companies, preferably with their latest weight (relative market capitalization) as well. I don't know where to get them easily so I'll just read them from the web html page (https://www.slickcharts.com/)

Normally the servers block such HTTP calls so we have to pretend that we are "human"


```
import pandas as pd
import requests
import numpy as np

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

url = 'https://www.slickcharts.com/sp500'
r = requests.get(url, headers=header)
sp500 = pd.read_html(r.text)
sp500 = sp500[0]
del sp500['#']

url = 'https://www.slickcharts.com/nasdaq100'
r = requests.get(url, headers=header)
nasdaq = pd.read_html(r.text)
nasdaq = nasdaq[0]
del nasdaq['#']

url = 'https://www.slickcharts.com/dowjones'
r = requests.get(url, headers=header)
dowjones = pd.read_html(r.text)
dowjones = dowjones[0]
del dowjones['#']
```

Let's display the dataframes.

```
sp500
```

Alright this is what we wanted! You can also display Nasdaq 100 and Dow Jones, but for now I'll stick to S&P500. 

As it says in the web page, there are more than 500 companies in the list due to some companies having more than one class of share. Let's try merge them.

I'll assume that the duplicate shares have almost the same company name.

```
duplicates = []

for company in sp500['Company'].tolist():
    for company_ in sp500['Company'].tolist():
        num_char_diff = sum(x!=y for x,y in zip(company, company_))
        if num_char_diff == 1:
            duplicates.append(company)

if len(duplicates) % 2 != 0:
    raise ValueError('something went wrong')

duplicates = sorted(duplicates)
duplicates = [(duplicates[2*i], duplicates[2*i+1])for i in range(len(duplicates)//2)]
duplicates
```

Seems about right!

From now on, I'll only keep Class A

```
duplicates = [company[1] for company in duplicates]

for index, row in sp500.iterrows():
    if row['Company'] in duplicates:
        sp500.drop(index, inplace=True)
```

```
sp500
```

Now finally we have 500 companies

Let's get their symbols

```
symbols500 = (sp500['Symbol']).tolist()
# somehow yahoo finance API works with '-' not with '.' for some symbols.
symbols500 = [symbol.replace('.', '-') for symbol in symbols500]
```

Let's get all of their data from yahoo finance

```
data = {symbol: yf.Ticker(symbol).history(period="1y", interval='1d')['Close']
        for symbol in symbols500}
```

Let's plot all of them. This might take some time.

```
import matplotlib.pyplot as plt
import mpld3
from mpld3 import plugins

plt.figure(figsize=(10,10))
for key, val in data.items():
    plt.plot(val)
mpld3.display()
```
